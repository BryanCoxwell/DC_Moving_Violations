# Moving Violations issued in Washington, DC

Data are sourced from opendata.dc.gov and cover the time between Jan 1st, 2021 and March 31st, 2022

```python
%matplotlib inline
import os
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

```python
data_files = glob.glob("data/*.csv")
moving_violations_dataset = pd.concat((pd.read_csv(f, low_memory=False) for f in data_files))
# Remove rows without X, Y coords for now
moving_violations_dataset.dropna(subset=['XCOORD', 'YCOORD'], how='all', inplace=True)
```

## Plot (XCOORD, YCOORD)
Straightforward visualizations of the XCOORD and YCOORD column

```python
x = moving_violations_dataset["XCOORD"]
y = moving_violations_dataset["YCOORD"]

FIG_WIDTH = 12
FIG_HEIGHT = 12

def draw(func, *args, **kwargs):
    plt.figure(figsize=(FIG_WIDTH, FIG_HEIGHT))
    func(*args, **kwargs)
    
print(len(x))
```

### Location Scatter Plot

```python
draw(plt.scatter, x, y, s=2)
```

### Location Histogram

```python
draw(plt.hexbin, x, y, gridsize=100, bins='log')
```

### Moving Violations Map By Issuing Agency

```python
moving_violations_dataset["ISSUING_AGENCY_SHORT"].unique()
```

```python
mpd_color = 'darkblue'
fed_color = 'red'
na_color = 'lightgreen'

color_dict = {
    'MPD-1D': mpd_color,
    'MPD-2D': mpd_color,
    'MPD-3D': mpd_color,
    'MPD-4D': mpd_color,
    'MPD-5D': mpd_color,
    'MPD-6D': mpd_color,
    'MPD-7D': mpd_color,
    'MPD-CID': mpd_color,
    'MPD-RC': mpd_color,
    'MPD-SOD': mpd_color,
    'MPD-PRTY': mpd_color,
    'MPD-YD': mpd_color,
    'MPD-OC': mpd_color,
    'MPD-SID': mpd_color,
    'MPD-SOMB': mpd_color,
    'MPD-DRD': mpd_color,
    'USPP': fed_color,
    'FBI': fed_color,
    'FPS': fed_color,
    'USCP': fed_color,
    'DCHA': fed_color,
    'DIA': fed_color,
    'DCTC': fed_color,
    'MTP': fed_color,
    'USSS': fed_color,
    'PSD': fed_color,
    'GPO': fed_color,
    'NZP': fed_color,
    'BEP': fed_color,
    'NO AGENCY': na_color,
    'OTHER': na_color,
}

plt.figure(figsize=(17, 17))
plt.axis('off')
plt.scatter(x, y, c=moving_violations_dataset['ISSUING_AGENCY_SHORT'].map(color_dict), s=10)
fake_handles = [mpatches.Patch(color=item) for item in color_dict.values()]
label = color_dict.keys()
plt.legend(fake_handles, label, loc='upper right', prop={'size': 8})
plt.savefig("by_agency.png")
```
