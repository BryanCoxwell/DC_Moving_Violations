# Moving Violations, Washington DC
## March 2022
Dataset sourced from DC Open Data

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

dataset_path = "./Datasets/Moving_Violations_Issued_in_March_2022.csv"
```

```python
moving_violations_dataset = pd.read_csv(dataset_path, low_memory = False)
```

```python
moving_violations_dataset.head()
```

#### Plot (XCOORD, YCOORD)
To make life easier I'll convert every value to an integer.

```python
def as_int(coord):
    return int(coord*1000)

x = moving_violations_dataset["XCOORD"].dropna().map(as_int)
y = moving_violations_dataset["YCOORD"].dropna().map(as_int)

FIG_WIDTH = 12
FIG_HEIGHT = 12

def draw(func, *args, **kwargs):
    plt.figure(figsize=(FIG_WIDTH, FIG_HEIGHT))
    func(*args, **kwargs)
```

#### Location Scatter Plot

```python
draw(plt.scatter, x, y)
```

#### Location Histogram

```python
draw(plt.hexbin, x, y, gridsize=50, bins='log')
```

#### Time Histogram

```python
issue_time = moving_violations_dataset["ISSUE_TIME"].dropna()
```

```python
draw(plt.hist, issue_time, bins=96)
```

#### State Plate

```python
states = moving_violations_dataset["PLATE_STATE"]
for state in states.unique():
    if state.strip():
        total = len(moving_violations_dataset[states == state])
        print("Totals for {0}: {1}".format(state, total))
    
```
